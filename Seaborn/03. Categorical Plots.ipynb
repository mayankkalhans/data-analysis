```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
titanic=sb.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.info()
```

## barplot()

```python
sb.barplot(x='sex', y='survived', data=titanic)
```

```python
sb.barplot(x='sex', y='survived', data=titanic, estimator=np.std)
```

```python
sb.barplot(x='sex', y='survived', data=titanic, hue='class')
```

```python
sb.barplot(x='sex', y='survived', data=titanic, hue='embarked')
```

```python
sb.barplot(x='sex', y='survived', data=titanic, hue='embarked', palette='dark')
```

## pointplot()

```python
sb.pointplot(x='sex', y='survived', data=titanic)
```

```python
sb.pointplot(x='sex', y='survived', data=titanic, estimator=np.std)
```

```python
sb.pointplot(x='sex', y='survived', data=titanic, hue='class')
```

```python
sb.pointplot(x='sex', y='survived', data=titanic, hue='embarked')
```

```python
sb.pointplot(x='sex', y='survived', data=titanic, hue='embarked', palette='dark')
```

## countplot()

```python
sb.countplot(x='sex', data=titanic)
```

```python
sb.countplot(x='sex', data=titanic, hue='alive')
```

```python
sb.countplot(x='sex', data=titanic, hue='alive', palette='Blues')
```

```python
sb.countplot(x='sex', data=titanic, hue='alive', palette='Blues_r')
```

```python
tips=sb.load_dataset('tips')
tips.head()
```

## box and whisker plot/boxplot()

```python
sb.boxplot(x='sex', y='total_bill', data=tips)
```

```python
sb.boxplot(x='day', y='total_bill', data=tips)
```

```python
sb.boxplot(x='day', y='total_bill', data=tips, palette='Greens')
```

```python
sb.boxplot(x='day', y='total_bill', data=tips, hue='smoker', palette='BrBG')
```

## violinplot()

```python
sb.violinplot(x='sex', y='total_bill', data=tips)
```

```python
sb.violinplot(x='day', y='total_bill', data=tips)
```

```python
sb.violinplot(x='day', y='total_bill', data=tips, palette='Greens')
```

```python
sb.violinplot(x='day', y='total_bill', data=tips, hue='sex', palette='BrBG')
```

```python
sb.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

## stripplot()

```python
sb.stripplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sb.stripplot(x='day', y='total_bill', data=tips)
```

```python
sb.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sb.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

## swarmplot()

```python
sb.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sb.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sb.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
sb.violinplot(x='day', y='total_bill', data=tips)
sb.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

## factorplot()

```python
sb.factorplot(x='day', y='total_bill', data=tips)
```

```python
sb.catplot(x='day', y='total_bill', data=tips)
```
