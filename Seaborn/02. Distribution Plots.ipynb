```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
iris=sb.load_dataset('iris')
```

```python
iris.head()
```

## distplot()

```python
sb.distplot(iris['sepal_length'])
```

```python
sb.distplot(iris['sepal_length'], kde=False)
```

```python
sb.distplot(iris['sepal_length'], hist=False)
```

```python
sb.distplot(iris['sepal_length'], bins=15)
```

```python
sb.distplot(iris['sepal_length'], bins=15, kde=False)
```

```python
sb.distplot(iris['sepal_length'], bins=15, rug=True)
```

```python
sb.distplot(iris['sepal_length'], bins=15, color='red')
```

```python
sb.distplot(iris['sepal_length'], bins=15, vertical=True)
```

## jointplot()

```python
sb.jointplot(x ='petal_length', y='petal_width', data=iris)
```

```python
sb.jointplot(x ='petal_length', y='petal_width', data=iris, kind='reg')
```

```python
sb.jointplot(x ='petal_length', y='petal_width', data=iris, kind='hex')
```

```python
sb.jointplot(x ='petal_length', y='petal_width', data=iris, kind='resid')
```

```python
sb.jointplot(x ='petal_length', y='petal_width', data=iris, kind='kde')
```

## pairplot()

```python
sb.pairplot(data=iris)
```

```python
sb.pairplot(data=iris, hue='species')
```

```python
sb.pairplot(data=iris, hue='species', palette='dark')
```

## rugplot()

```python
sb.rugplot(iris['petal_length'])
```

```python
sb.rugplot(iris['petal_length'], axis='y')
```

```python
sb.kdeplot(data=iris['sepal_length'], shade=True, cumulative=True)
```
