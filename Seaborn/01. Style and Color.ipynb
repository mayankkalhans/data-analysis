## Importing Datasets as Pandas Dataframe and Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
tips=sb.load_dataset('tips')
tips.head()
```

## Figure Aesthetic

#### Plotting using matplotlib

```python
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x, np.sin(x+1*0.5)*(7-i)*flip)

sinplot()
```

#### Plotting using seaborn

```python
sb.set()
sinplot()
```

#### Seaborn Figure Styles

```python
sb.set_style('whitegrid')
sinplot()
```

```python
sb.set_style('ticks')
sinplot()
```

```python
sb.set(style='whitegrid')
sinplot()
```

#### Removing Axes Spines

```python
sb.set(style='whitegrid')
sinplot()
sb.despine()
```

```python
sb.set(style='whitegrid')
sinplot()
sb.despine(left=True)
```

#### Overriding the Elements

```python
sb.axes_style()
```

```python
sb.set_style('whitegrid', {'axes.axisbelow':False})
sinplot()
sb.despine()
```

## Color Palette

#### Qualitative Color Palette
Qualitative or categorical palettes are best suitable to plot the categorical data.

```python
current_palette=sb.color_palette()
sb.palplot(current_palette)
```

```python
current_palette=sb.color_palette(palette='deep')
sb.palplot(current_palette)
```

```python
current_palette=sb.color_palette(palette='muted')
sb.palplot(current_palette)
```

```python
current_palette=sb.color_palette(palette='bright', n_colors=6)
sb.palplot(current_palette)
```

```python
current_palette=sb.color_palette(palette='pastel', n_colors=14)
sb.palplot(current_palette)
```

```python
current_palette=sb.color_palette(palette='dark')
sb.palplot(current_palette)
```

```python
current_palette=sb.color_palette(palette='colorblind', n_colors=4)
sb.palplot(current_palette)
```

#### Sequential Color Palette
Sequential palettes are suitable to express the distribution of data ranging from relative lower values to higher values with a range.
Appending an additional character 's' to the color passed to the color parameter will plot the sequential plot.

```python
sb.palplot(sb.color_palette('Greens'))
```

```python
sb.palplot(sb.color_palette('dark'))
```

```python
sb.palplot(sb.color_palette('Blues'))
```

```python
sb.palplot(sb.color_palette('Blues_r'))
```

```python
sb.palplot(sb.color_palette('Blues_d'))
```

```python
sb.palplot(sb.color_palette('Greens_r'))
```

```python
sb.palplot(sb.color_palette('Greens_d'))
```

#### Diverging Color Palette
Diverging palettes use two different colors. Each color represents variation in the value ranging from a common point in either direction.

```python
sb.palplot(sb.color_palette('BrBG',7))
```

```python
sb.palplot(sb.color_palette('BuGn',7))
```

```python
sb.set(style='white', palette='dark')
sinplot()
```
