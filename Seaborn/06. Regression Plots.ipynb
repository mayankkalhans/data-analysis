```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sb.load_dataset('tips')
tips.head()
```

```python
anscombe=sb.load_dataset('anscombe')
anscombe.head()
```

## regplot()

```python
sb.regplot(x='total_bill', y='tip', data=tips)
```

## lmplot()

```python
sb.lmplot(x='total_bill', y='tip', data=tips)
```

You can see the difference in the size between the two plots 

```python
sb.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sb.lmplot(x='size', y='tip', data=tips)
```

#### Working with Markers

```python
sb.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```

```python
sb.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', '1'])
```

```python
sb.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', '1'], scatter_kws={'s':100})
```

#### Using a Grid

```python
sb.lmplot(x='total_bill', y='tip', col='sex', data=tips)
```

```python
sb.lmplot(x='total_bill', y='tip', row='sex',col='time', data=tips)
```

```python
sb.lmplot(x='total_bill', y='tip', col='day', data=tips, hue='sex')
```

#### Size and Aspect

```python
sb.lmplot(x='total_bill', y='tip', col='day', data=tips, hue='sex', aspect=0.6, height=8)
```

#### Fitting Different Kinds of Models

```python
sb.lmplot(x='x', y='y', data=anscombe.query("dataset=='I'"))
```

```python
sb.lmplot(x='x', y='y', data=anscombe[anscombe['dataset']=='I'])
```

```python
sb.lmplot(x='x', y='y', data=anscombe.query("dataset=='II'"), order=2)
```
