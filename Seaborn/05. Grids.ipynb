```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
iris=sb.load_dataset('iris')
iris.head()
```

```python
tips=sb.load_dataset('tips')
tips.head()
```

## PairGrid()

```python
sb.PairGrid(iris)
```

```python
g=sb.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sb.PairGrid(iris, hue='species')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

```python
g=sb.PairGrid(iris, hue='species')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter).add_legend()
```

```python
g = sb.PairGrid(iris, hue='species')
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sb.kdeplot, legend=True)
```

```python
g = sb.PairGrid(iris, hue='species')
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sb.kdeplot, legend=False)
```

## FacetGrid()

```python
g=sb.FacetGrid(data=tips, col='time', row='smoker')
```

```python
g = sb.FacetGrid(tips, col="time",  row="smoker")
g.map(plt.hist, "total_bill")
```

```python
g = sb.FacetGrid(tips, col="time",  row="smoker", hue='sex')
g.map(plt.scatter, "total_bill", 'tip')
```

```python
g = sb.FacetGrid(tips, col="time",  row="smoker", hue='sex')
g.map(plt.scatter, "total_bill", 'tip').add_legend()
```

## JointGrid()

```python
g = sb.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sb.JointGrid(x="total_bill", y="tip", data=tips)
g.plot(sb.regplot, sb.distplot)
```
