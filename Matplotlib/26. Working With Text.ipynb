```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.set_title('axes title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.text(x=3, y=8, s='boxed italics text in data coords', style='italic',bbox={'facecolor':'red'})
ax.text(2,6, r'an equation: $E=mc^2$', fontsize=15)
ax.text(4,0.05, 'COLORED TEXT IN AXES CORD', verticalalignment='bottom', color='green', fontsize=15)
ax.plot([2],[1], marker='o')
ax.annotate('annotate', xy=(2,1), xytext=(2,4), arrowprops=dict(facecolor='black', shrink=0.05))
ax.axis([0,10,0,10])
```
