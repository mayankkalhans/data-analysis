```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tk
import matplotlib
%matplotlib inline
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
x = np.arange(1,5)
axes[0].plot(x, np.exp(x))
axes[0].plot(x, x**2)
axes[0].set_title('Normal Scale')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')
axes[0].xaxis.labelpad = 10
axes[0].yaxis.labelpad = 10

axes[1].plot(x, np.exp(x))
axes[1].plot(x, x**2)
axes[1].set_yscale('log')
axes[1].set_title('Logarithmic scale (y)')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')

```

## Axis Spines

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
ax.plot([1,2,3,4,5], lw = 2)
```

## Scientific Notation

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, x, np.exp(x))
ax.set_title('Scientific Notation')
ax.set_yticks([0, 50, 100, 150])

formatter = tk.ScalarFormatter(useMathText = True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1, 1))
ax.yaxis.set_major_formatter(formatter)
```

## Axis Number and Axis Label Spacing

```python
#distance between numbers on the axes
matplotlib.rcParams['xtick.major.pad']=5
matplotlib.rcParams['ytick.major.pad']=5

fig,ax = plt.subplots(1,1)
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0,50,100,150])

#padding between axis label and axis numbers
ax.xaxis.labelpad=5
ax.yaxis.labelpad=5

ax.set_xlabel('x')
ax.set_ylabel('y')
```

## Axes where x and y is zero

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color(None)
ax.spines['top'].set_color('None')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

y = np.linspace(-0.75, 1., 100)
ax.plot(y, y**3);
```

## Text Annotation

```python
fig, ax = plt.subplots()

ax.plot(y, y**2, y, y**3)

ax.text(0.15, 0.2, r"$y=x^2$", fontsize=20, color="blue")
ax.text(0.65, 0.1, r"$y=x^3$", fontsize=20, color="green", alpha=0.5);
```
