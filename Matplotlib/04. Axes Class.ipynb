```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y = [1, 4, 9, 14, 25, 36, 49, 64]
x1 = [1, 16, 30, 42, 55, 68, 77, 88]
x2 = [1, 6, 12, 18, 28, 48, 52, 65]
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1, y, 'rd--')
l2 = ax.plot(x2, y, 'mo-')
ax.legend(labels = ('TV', 'Smartohone'), loc = 0 )
ax.set_title('Advertisement effect on sales')
ax.set_xlabel('Medium')
ax.set_ylabel('Sales')
```
