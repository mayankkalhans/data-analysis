```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1, 10)
a1.plot(x, np.exp(x))
a1.set_title('Without Limits')
```

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1, 10)
a1.plot(x, np.exp(x))
a1.set_title('With Limits')
a1.set_xlim(0,10)
a1.set_ylim(0, 16000)
```
