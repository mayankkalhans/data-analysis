```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')

langs=['C++', 'C', 'R', 'Java', 'Python']
students=[54, 48, 84, 75, 102]

ax.pie(students, labels=langs, autopct='%1.2f%%')
```
