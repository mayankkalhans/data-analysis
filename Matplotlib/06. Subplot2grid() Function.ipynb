```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a1 = plt.subplot2grid((3,3), (0,0), colspan = 2)
a2 = plt.subplot2grid((3,3), (0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3), (1,0), rowspan = 2, colspan = 2)
x = np.arange(1, 10)
a2.plot(x, x*x)
a2.set_title('Square')
a1.plot(x, np.exp(x))
a1.set_title('Exponential')
a3.plot(x, np.log10(x))
a3.set_title('Log')
```

```python
fig = plt.figure()
a1 = fig.subplot2grid((3,3), (0,0), colspan = 2)
a2 = fig.subplot2grid((3,3), (0,2), rowspan = 3)
a3 = fig.subplot2grid((3,3), (1,0), rowspan = 2, colspan = 2)
x = np.arange(1, 10)
a2.plot(x, x*x)
a2.set_title('Square')
a1.plot(x, np.exp(x))
a1.set_title('Exponential')
a3.plot(x, np.log10(x))
a3.set_title('Log')
```
