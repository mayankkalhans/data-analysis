```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
x = np.arange(0, math.pi*2, 0.05)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, np.sin(x))
ax.set_xlabel('Angle')
ax.set_ylabel('Sine')
```

```python
x = np.arange(0, math.pi*2, 0.05)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, np.sin(x))
ax.set_xlabel('Angle')
ax.set_ylabel('Sine')
ax.set_xticks([0,2,4,6,8])
```

```python
x = np.arange(0, math.pi*2, 0.05)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, np.sin(x))
ax.set_xlabel('Angle')
ax.set_ylabel('Sine')
ax.set_xticks([0,2,4,6])
ax.set_yticks([-1,0,1])
ax.set_xticklabels(['zero', 'two', 'four', 'six'])
```
