```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import math
```

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('sine wave')
ax.set_xlabel('angle')
ax.set_ylabel('sine')
ax.plot(x,y)
```
