```python
x = np.arange(0, math.pi*2, 0.05)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.55, 0.55, 0.3, 0.3])
ax1.plot(x, np.sin(x), 'r--')
ax2.plot(x, np.cos(x), 'b-.')
ax1.set_title('Sine')
ax2.set_title('Cosine')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 1, 1])
ax2 = fig.add_axes([0.55, 0.55, 0.3, 0.3])
ax1.plot(x, np.sin(x), 'r--')
ax2.plot(x, np.cos(x), 'b-.')
ax1.set_title('Sine')
ax2.set_title('Cosine')
```
