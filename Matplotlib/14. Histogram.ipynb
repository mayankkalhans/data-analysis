```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig,ax=plt.subplots()
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins=[0,25,50,75,100])
ax.set_xticks([0,25,50,75,100])
ax.set_title('Histogram')
ax.set_xlabel("Marks")
ax.set_ylabel('Number of students')
```

```python
fig,ax=plt.subplots()
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins=4, range=(0,100))
ax.set_xticks([0,25,50,75,100])
ax.set_title('Histogram')
ax.set_xlabel("Marks")
ax.set_ylabel('Number of students')
```

```python
fig,ax=plt.subplots()
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins=4, range=(0,100), cumulative=True)
ax.set_xticks([0,25,50,75,100])
ax.set_title('Histogram')
ax.set_xlabel("Marks")
ax.set_ylabel('Number of students')
```

```python
fig,ax=plt.subplots()
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins=4, range=(0,100))
ax.set_xticks([0,25,50,75,100])
ax.set_title('Histogram')
ax.set_xlabel("Marks")
ax.set_ylabel('Number of students')
```
