```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.outer(np.linspace(-2,2,30), np.ones(30))
y=x.copy().T
z=np.cos(x**2+y**2)

fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z, cmap='viridis', edgecolor='black')
plt.tight_layout()
```
