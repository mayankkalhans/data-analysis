```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure(figsize=(8, 2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(6, 3))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,2), dpi = 100)
axes[0][0].plot(x,y)
axes[1][1].plot(y,x)
plt.tight_layout()
```
