```python
fig = plt.figure()
a = fig.subplots(2, 2)
x = np.arange(1, 5)
a[0][0].plot(x, x*x)
a[0][0].set_title('Square')
a[0][1].plot(x, np.sqrt(x))
a[0][0].set_title('Square Root')
a[1][0].plot(x, np.exp(x))
a[1][0].set_title("Exponential")
a[1][1].plot(x, np.log10(x))
a[1][1].set_title('Log')
```

```python
plt.subplot(221)
plt.plot(x, x*x)
plt.title('Square')
plt.subplot(222)
plt.plot(x, np.sqrt(x))
plt.title('Square root')
plt.subplot(223)
plt.plot(x, np.exp(x))
plt.title('Exponential')
plt.subplot(224)
plt.plot(x, np.log10(x))
plt.title('Log')
```

```python
fig = plt.figure()
x = np.arange(1, 5)
ax1 = fig.add_subplot(2,2,1)
ax1.plot(x, x*x)
ax1.set_title('Square')
ax2 = fig.add_subplot(2,2,2)
ax2.plot(x, np.sqrt(x))
ax2.set_title('Square Root')
ax3 = fig.add_subplot(2,2,3)
ax3.plot(x, np.exp(x))
ax3.set_title('Exponential')
ax4 = fig.add_subplot(2,2,4)
ax4.plot(x, np.log10(x))
ax4.set_title('Log')
```
