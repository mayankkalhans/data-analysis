```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(10)
collection_1=np.random.normal(100,10,200)
collection_2=np.random.normal(80,30,200)
collection_3=np.random.normal(90,20,200)
collection_4=np.random.normal(70,25,200)
data_to_plot=[collection_1,collection_2,collection_3,collection_4]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.violinplot(data_to_plot)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.violinplot(data_to_plot, showmeans=True)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.violinplot(data_to_plot, showmedians=True,showmeans=True)
```
