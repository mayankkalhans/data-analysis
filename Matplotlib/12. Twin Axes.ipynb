```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(1, 11)
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
a1.plot(x, np.exp(x), label ='exp')
a1.set_ylabel('Exponential')
a1.set_xlabel('Values')
a2 = a1.twinx()
a2.plot(x, np.log(x), color='red', marker='o', label='log')
a2.set_ylabel('Log')
fig.legend(loc = 'upper left')
```

```python
x = np.arange(1, 11)
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
a1.plot(x, np.exp(x), label ='exp')
a1.set_ylabel('Exponential')
a2 = a1.twinx()
a2.plot(x, np.log(x), color='red', marker='o', label='log')
a2.set_ylabel('Log')
a1.set_xlabel('Values')
a1.legend(loc = 'upper left')
a2.legend(loc = 'lower right')
```

```python
x = np.arange(1, 11)
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
a1.plot(x, np.exp(x))
a1.set_ylabel('Exponential')
a2 = a1.twinx()
a2.plot(x, np.log(x), color='red', marker='o')
a2.set_ylabel('Log')
fig.legend(labels=('exp', 'log'), loc = 'upper left')
```
