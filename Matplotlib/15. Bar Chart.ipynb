```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Simple Bar Graph

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

langs=['C++', 'C', 'R', 'Java', 'Python']
students=[54, 48, 84, 75, 102]

axes.bar(x=langs, height=students, width=0.8, align='center')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

langs=['C++', 'C', 'R', 'Java', 'Python']
students=[54, 48, 84, 75, 102]

axes.bar(x=langs, height=students, width=0.8, align='edge')
```

## Horizontal Bar Graph

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

langs=['C++', 'C', 'R', 'Java', 'Python']
students=[54, 48, 84, 75, 102]

axes.barh(y=langs, width=students, height=0.8)
```

## Multi-Bar Bar Graph

```python
data=[[30,25,40,20],[45,60,25,30],[60,50,45,35]]
X=np.arange(4)

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(X+0.00, data[0], width=0.25)
ax.bar(X+0.25, data[1], width=0.25)
ax.bar(X+0.50, data[2], width=0.25)
ax.set_xticks([0.25, 1.25, 2.25, 3.35])
ax.set_xticklabels(['2015', '2016', '2017', '2018'])
ax.legend(labels=('CS', 'IT', 'EC'))
```

## Stacked Bar Graph

```python
menmean=[30,40,50,60]
womenmean=[20,30,10,50]
x=np.arange(4)

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(x, menmean, width=0.35)
ax.bar(x, womenmean, width=0.35, bottom=menmean)
ax.set_ylabel('Scores')
ax.set_title('Scores by Gender')
ax.set_xticks([0,1,2,3])
ax.set_xticklabels(['G1', 'G2', 'G3', 'G4'])
ax.legend(labels=('Men', 'Women'))
```
