```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x,y=np.meshgrid(np.arange(-2,2,0.2), np.arange(-2,2,0.25))
z=x*np.exp(-x**2-y**2)
v,u=np.gradient(z,0.2,0.2)
fig,ax=plt.subplots()
q=ax.quiver(x,y,u,v)
```

```python
x,y=np.meshgrid(np.arange(-2,2,0.2), np.arange(-2,2,0.25))
z=x*np.exp(-x**2-y**2)
v,u=np.gradient(z,0.2,0.2)
fig,ax=plt.subplots()
q=ax.quiver(x,y,u,v, color='blue')
```
