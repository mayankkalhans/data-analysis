```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
xlist=np.linspace(-3,3,100)
ylist=np.linspace(-3,3,100)
x,y=np.meshgrid(xlist, ylist)
z=np.sqrt(x**2+y**2)
```

```python
fig,ax=plt.subplots(1,1)
cp=ax.contourf(x,y,z)
fig.colorbar(cp)
ax.set_xlabel('x(cm)')
ax.set_ylabel('y(cm)')
```

```python
fig,ax=plt.subplots(1,1)
cp=ax.contourf(x,y,z, levels=5)
fig.colorbar(cp)
ax.set_xlabel('x(cm)')
ax.set_ylabel('y(cm)')
```

```python
fig,ax=plt.subplots(1,1)
cp=ax.contour(x,y,z)
fig.colorbar(cp)
ax.set_xlabel('x(cm)')
ax.set_ylabel('y(cm)')
```

```python
fig,ax=plt.subplots(1,1)
cp=ax.contour(x,y,z, levels=8)
fig.colorbar(cp)
ax.set_xlabel('x(cm)')
ax.set_ylabel('y(cm)')
```
