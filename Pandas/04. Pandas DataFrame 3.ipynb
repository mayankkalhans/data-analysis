## nlargest() and nsmallest() Method
Returns the n largest and n smallest values respectively in a column. For a large number of data it"s better to use nlargest() and nsmallest() Method as they are faster than sort_values() Method.

```python
bond.nlargest(n=5, columns='Box Office')
```

```python
bond['Box Office'].nlargest(7)
```

```python
bond.nsmallest(n=6, columns='Box Office')
```

```python
bond['Box Office'].nsmallest()
```

## apply() Method

#### Using apply() on columns

```python
def add_millions(num):
    return str(num) + ' Millions'
```

```python
bond['Box Office'].apply(add_millions)
```

#### Using apply() on rows

```python
 def movie_rating(row):
        actor = row[1]
        budget = row[4]
        if actor=='Pierce Brosnan':
            return 'Outstanding'
        elif actor=='Roger Moore' and budget > 40:
            return 'Enjoyable'
        else:
            return 'I have no clue'
```

```python
bond.apply(movie_rating, axis=1)
```
