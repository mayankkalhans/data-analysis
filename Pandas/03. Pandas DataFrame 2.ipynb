```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management']=df['Senior Management'].astype('bool')
df['Gender']=df['Gender'].astype('category')
```

```python
df
```

```python
df.info()
```

## Filter a Dataframe based on a Condition

```python
mask = df['Team']=='Finance'
df[mask]
```

```python
mask = df['Salary']>=100000
df[mask]
```

## Filter with more than one Condition

#### Using &(AND)

```python
mask1 = df['Team'] == 'Finance'
mask2 = df['Salary'] >= 100000
df[mask1 & mask2]
```

#### Using |(OR)

```python
mask1 = df['Team'] == 'Finance'
mask2 = df['Salary'] >= 100000
df[mask1 | mask2]
```

#### Using both & and |

```python
mask1 = df['Team'] == 'Finance'
mask2 = df['Salary'] >= 100000
mask3 = df['Start Date'] > '1985-01-01'
df[(mask1 & mask2) | mask3]
```

#### isin() Method
Used to check multiple values in a column to avoid creation of multiple commands 

```python
mask = df['Team'].isin(['Legal', 'Finance', 'Marketing'])
df[mask]
```

#### isnull() and notnull() Methods

```python
mask = df['Gender'].isnull()
df[mask]
```

```python
mask = df['Gender'].notnull()
df[mask]
```

#### between() Method

```python
df[df['Last Login Time'].between('08:30AM','01:00PM')]
```

```python
df[df['Last Login Time'].between('08:30:00','13:00:00')]
```

#### duplicated() Method

```python
df.sort_values('First Name', inplace=True) 
df
```

```python
mask = df['First Name'].duplicated(keep='first')
mask
```

```python
df[mask]
```

```python
mask = df.duplicated(subset=['First Name', 'Team'], keep='last')
df[mask]
```

```python
mask = df['First Name'].duplicated(keep=False)
df[mask]
```

```python
mask = ~df['First Name'].duplicated(keep=False)
df[mask]
```

#### drop_duplicates() Method

```python
df.drop_duplicates(subset=['First Name'])
```

```python
df.drop_duplicates(subset=['First Name'],  keep='last')
```

```python
df.drop_duplicates(subset=['First Name'],  keep=False)
```
