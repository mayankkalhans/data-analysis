```python
import pandas as pd
import numpy as np
```

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
week1_satisfaction = pd.read_csv('Restaurant - Week 1 Satisfaction.csv')
```

## concat() Method

```python
pd.concat([week1, week2])
```

```python
pd.concat([week1, week2], ignore_index=True)
```

```python
sales = pd.concat([week1, week2], keys=['Week 1', 'Week 2'])
```

```python
sales
```

```python
sales.loc['Week 2']
```

```python
sales.loc[('Week 2', 240)]
```

```python
sales.loc[('Week 2', 240), 'Customer ID']
```

## append() Method

```python
week1.append(week2)
```

```python
week2.append(week1)
```

```python
week1.append(week2, ignore_index=True)
```

```python
week1.append([week2, week1], ignore_index=True)
```

## Merging/Joining

#### Inner Join

```python
week1.merge(week2, on = 'Customer ID', how='inner')
```

```python
week2.merge(week1, on = 'Customer ID', how='inner')
```

```python
week1.merge(week2, on = 'Customer ID', how='inner', suffixes=['- Week 1', '- Week 2'])
```

```python
week1.merge(week2, how = 'inner', on = ['Customer ID', 'Food ID'])
```

#### Outer Join

```python
week1.merge(week2, on = 'Customer ID', how='outer', suffixes=['- Week 1', '- Week 2'])
```

```python
merged=week1.merge(week2, on = 'Customer ID', how='outer', suffixes=['- Week 1', '- Week 2'], indicator=True)
```

```python
merged
```

```python
mask = merged['_merge'].isin(['left_only', 'right_only'])
```

```python
merged[mask]
```

#### Left Join

```python
week1.merge(foods, how = 'left', on = 'Food ID')
```

```python
week1.merge(foods, how = 'left', on = 'Food ID', sort = True)
```

#### left_on and right_on

```python
week2.merge(customers, how = 'left', left_on='Customer ID', right_on='ID')
```

```python
week2.merge(customers, how = 'left', left_on='Customer ID', right_on='ID').drop('ID', axis=1)
```

#### left_index and right_index

```python
week1.merge(week2, how='left', left_index=True, right_index=True, suffixes=[' - Week 1', ' - Week 2'])
```

#### join() method

```python
week1.merge(week1_satisfaction, how='left', left_index=True, right_index=True)
```

```python
week1.join(week1_satisfaction)
```
