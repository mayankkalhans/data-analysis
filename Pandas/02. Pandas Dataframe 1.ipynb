## Broadcasting Operations
The operation is applied on every data of the column

```python
print((nba['Age'] + 5).head())
#Also written as nba['Age'].add(5)
print(nba['Age'].add(5).head())
```

```python
print((nba['Age'] - 7).head())
#Also written as nba['Age'].sub(7)
print(nba['Age'].sub(7).head())
```

```python
print((nba['Weight'] * 0.42).head())
#Also written as nba['Weight'].mul(0.42)
print(nba['Weight'].mul(0.42).head())
```

```python
print((nba['Salary']/100000).head())
#Also written as nba['Salary'].div(100000)
print(nba['Salary'].div(100000).head())
```

## Drop Rows with Null Values using dropna() Method

#### Drop all rows having NaN in any column

```python
nba.dropna()
```

#### Drop rows having NaN in all columns

```python
nba.dropna(how='all', inplace=True)
nba
```

#### Drop columns having NaN

```python
nba.dropna(axis=1)
```

#### Drop row from a particular column

```python
nba.dropna(subset=['Salary'])
```

## Fill in Missing Values using fillna() Method

```python
nba['Salary'].fillna(0, inplace = True)
nba
```

```python
nba['College'].fillna('No College', inplace = True)
nba
```

## astype() Method

```python
nba['Salary'] = nba['Salary'].astype('int')
nba['Position'] = nba['Position'].astype('category')
nba['Team'] = nba['Team'].astype('category')
```

```python
nba.info()
```

## Sorting a Dataframe

#### sort_values() Method

sorting a single column

```python
nba.sort_values('Name')
```

sorting more than one column, one in ascending order and the other in descending order

```python
nba.sort_values(['Age', 'Name'], ascending=[True, False], inplace=True)
```

```python
nba
```

sorting by index

```python
nba.sort_index(inplace=True)
```

```python
nba
```

sort by axis

```python
nba.sort_index(axis=1)
```

## rank() Method
Returns a rank of every respective index of a series passed.

```python
nba['Name Rank']=nba['Name'].rank()
```

```python
nba.sort_values('Name', inplace=True)
```

```python
nba.head(36)
```

```python
del nba['Name Rank']
```
