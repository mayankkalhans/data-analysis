```python
import pandas as pd
import numpy as np
```

```python
chicago = pd.read_csv('chicago.csv').dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
chicago.head(10)
```

```python
chicago.info()
```

## Different String Methods

#### title() Method

```python
chicago['Name'].str.title()
```

```python
chicago['Position Title'].str.title()
```

#### len() Method
If we don't use '.str' before len then it will give number of rows in dataframe.

```python
chicago['Department'].str.len()
```

```python
len(chicago)
```

#### replace() Method

```python
chicago['Department'].str.replace('MGMNT', 'MANAGEMENT')
```

```python
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].str.replace('$', '').astype(float)
```

```python
chicago
```

#### strip() Method

```python
chicago['Name'] = chicago['Name'].str.strip()
```

```python
chicago['Position Title'] = chicago['Position Title'].str.strip()
```

```python
chicago['Department'] = chicago['Department'].str.strip()
```

```python
chicago
```

## Filtering with String Methods

#### contains() Method

```python
mask = chicago['Position Title'].str.lower().str.contains('water')
chicago[mask]
```

####  startswith() Method

```python
mask = chicago['Position Title'].str.lower().str.startswith('water')
chicago[mask]
```

```python
mask = chicago['Position Title'].str.lower().str.endswith('ist')
chicago[mask]
```
