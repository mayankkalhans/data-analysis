## Bar Charts

```python
google = data.DataReader(name = 'GOOG', data_source = 'yahoo', start = '2004-01-01', end = '2020-5-2')
google.head()
```

```python
def stock_performance(stock_price):
    if stock_price <= 200:
        return 'Poor'
    elif stock_price >200 and stock_price <= 500:
        return 'Satisfactory'
    else:
        return 'Stellar'
```

```python
google['Close'].apply(stock_performance)
```

```python
goog = google['Close'].apply(stock_performance).value_counts()
```

```python
plt.style.use('ggplot')
goog.plot(kind='bar')
```

```python
google['Close'].apply(stock_performance).value_counts().plot()
```

```python
google['Close'].apply(stock_performance).value_counts().plot(kind = 'bar')
```

```python
google['Close'].apply(stock_performance).value_counts().plot(kind = 'barh')
```

## Pie Charts

```python
apple = data.DataReader(name = 'AAPL', data_source = 'yahoo', start = '2012-01-01', end = '2020-5-2')
apple.head()
```

```python
apple['Close'].mean()
```

```python
def stock_performance(stock_price):
    if stock_price >= 133.30992497014634:
        return 'Above Average'
    else:
        return 'Below Average'
```

```python
plt.style.use('ggplot')
apple['Close'].apply(stock_performance).value_counts().plot(kind = 'pie', legend = True)
```

## Histograms

```python
google = data.DataReader(name = 'GOOG', data_source = 'yahoo', start = '2004-01-01', end = '2020-5-2')
google.head()
```

```python
def custom_round(stock_price):
    return int(stock_price/100)*100
```

```python
google['High'].apply(custom_round).value_counts().sort_index()
```

```python
google['High'].apply(custom_round).nunique()
```

```python
google['High'].apply(custom_round).plot(kind = 'hist')
```

```python
google['High'].apply(custom_round).plot(kind = 'hist', bins = 16)
```
